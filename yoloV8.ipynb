```python
from ultralytics import YOLO
from IPython.display import Image
import cv2

# Load the YOLO model
model = YOLO("/content/gdrive/MyDrive/yolov8/runs/runs/detect/train2/weights/last.pt")

# Perform object detection on the image
results = model.predict(source="/content/gdrive/MyDrive/yolov8/test_2.jpeg")

# Display the image with detection results
image = cv2.imread("/content/gdrive/MyDrive/yolov8/test_2.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Draw bounding boxes and labels on the image
for result in results:
    for box in result.boxes:
        x_min, y_min, x_max, y_max = box.xyxy[0]
        x_min, y_min, x_max, y_max = int(x_min), int(y_min), int(x_max), int(y_max)
        label = model.names[int(box.cls[0])]
        confidence = float(box.conf[0])
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 0, 255), 2)
        cv2.putText(image, f"{label} {confidence:.2f}", (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)

# Display the image with detection results
Image(data=cv2.imencode('.jpg', image)[1].tobytes())
```

```python
from sahi import AutoDetectionModel
from sahi.predict import get_sliced_prediction
from PIL import Image

from sahi.utils.cv import visualize_object_predictions

import cv2
import numpy as np
from IPython.display import Image

detection_model = AutoDetectionModel.from_pretrained(
    model_type='yolov8',
    model_path="/content/gdrive/MyDrive/yolov8/runs/runs/detect/train2/weights/last.pt",
    confidence_threshold=0.35,
    device="cpu"
)



result = get_sliced_prediction(
   "/content/gdrive/MyDrive/yolov8/test_image11.jpg",
    detection_model=detection_model,
    slice_height=150,
    slice_width=150,
    overlap_height_ratio=0.2,
    overlap_width_ratio=0.2
)

# result.export_visuals(export_dir='/content/gdrive/MyDrive/yolov8/output_pic')
# Image('/content/gdrive/MyDrive/yolov8/test_image4.jpeg')


img = cv2.imread("/content/gdrive/MyDrive/yolov8/test_image11.jpg", cv2.IMREAD_UNCHANGED)
img_converted = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
numpydata = np.asarray(img_converted)
visualize_object_predictions(
    numpydata,
    object_prediction_list = result.object_prediction_list,
    hide_labels = False,
    output_dir='/content/gdrive/MyDrive/yolov8/output_pic',
    file_name = 'result',
    export_format = 'jpg',

)
Image('/content/gdrive/MyDrive/yolov8/output_pic/result.jpg')
```
